```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
plt.rcParams['figure.figsize'] = [7, 5]
plt.rcParams['figure.dpi'] = 120
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

```python
my_palette = ['Salmon', 'grey', 'Turquoise', 'royalblue']
sns.swarmplot(data = tips, 
              y = 'total_bill',
              x = 'day',
              hue = 'day',
              palette = my_palette
             );
```

```python
sns.swarmplot(data = tips, 
              y = 'total_bill',
              x = 'day',
              hue = 'sex'
             );
```

```python
# dodge
sns.swarmplot(data = tips, 
              y = 'total_bill',
              x = 'day',
              hue = 'sex',
              dodge = True
             );
```

```python
# size 
my_palette = ['Salmon', 'grey', 'Turquoise', 'royalblue', 'Orchid', 'SandyBrown']

sns.swarmplot(data=tips, x='day', y='tip',
              hue='size',
              palette=my_palette,
              dodge=True,
              size = 3
             );
```

```python
################################### swarmplot with boxplot ##########################################
my_palette = ['Salmon', 'grey', 'Turquoise', 'royalblue']

sns.boxplot(data=tips, x='day', y='total_bill',
            color='Cornsilk');

sns.swarmplot(data=tips, x='day', y='total_bill',
              hue='day',
              palette=my_palette);
```

```python
############### swarmplot with violinplot #################################
sns.violinplot(data=tips, x='day', y='total_bill',
            color='Cornsilk', inner=None);

sns.swarmplot(data=tips, x='day', y='total_bill',
              hue='day',
              palette=my_palette);
```

```python
# linewidth, edgecolor etc.
sns.swarmplot(data=tips, y='total_bill', 
              linewidth=2,
              color='white',
              edgecolor='coral',
             marker = '>'
             );
# 'o': круг (по умолчанию);
# 's': квадрат;
# 'd': ромб;
# 'v': треугольник вниз;
# '^': треугольник вверх;
# '>': треугольник вправо;
# '<': треугольник влево;
# 'p': пятиугольник;
# '*': звезда;
# 'h': шестигранник;
# '+': плюс.
```
