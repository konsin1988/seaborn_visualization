```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

plt.rcParams['figure.figsize'] = [7,5]
plt.rcParams['figure.dpi'] = 120
```

```python
df = sns.load_dataset('mpg', dtype={'cylinders': 'category', 'origin': 'category'}).dropna()
df.head()
```

```python
df.info()
```

```python
sns.heatmap(
    df
        .select_dtypes(['number'])
        .corr(),
    annot = True,
    cmap = 'GnBu',
    vmin=-1,
    vmax=1
);
```

```python
#  mask

#  mask  sns.heatmap()        . 
#         , , 
#         .

#  mask    ( DataFrame)   ,   , 
#   sns.heatmap().     True   , 
#    ,   False -  ,    .

#  ,   ,        . 
#          NumPy - np.eye(), 
#          True.

mask = np.eye(
    df
        .select_dtypes(['number'])
        .corr().shape[0], 
    dtype = bool
)
mask
```

```python
sns.heatmap(
    df
        .select_dtypes(['number'])
        .corr(),
    annot = True,
    cmap = 'GnBu',
    vmin=-1,
    vmax=1,
    mask = mask,
    linewidths = 1.5,
    linecolor = '#8dc2f0'
);
```
